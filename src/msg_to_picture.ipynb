### imports and setup

```python
# from pydrive.auth import GoogleAuth
# from pydrive.drive import GoogleDrive
import os
import extract_msg
import sys
```

```python
input_folder = "./../.data/"
output_folder = "./../.bin/pictures_in_png/"

if not os.path.exists(input_folder):
    os.makedirs(input_folder)
```

### connect and download files from google drive

```python
gauth = GoogleAuth()
drive = GoogleDrive(gauth)

# getting list of files in the folder
folder_id = '<folder_id>'
file_list = drive.ListFile({'q': f"'{folder_id}' in parents"}).GetList()

# downloading files
for file in file_list:
    try:
        print(f"Downloading file {file['title']}...")
        output = os.path.join("../../data/google_drive/raw/", file['title'])
        file.GetContentFile(output)
    except Exception as e:
        print(f"Could not download file {file['title']}. Reason: {e}")
```

### convert msg to png and save picture

```python
# iterating through the input folder for MSG files
for file in os.listdir(input_folder):
    if file.endswith(".msg"):
        input_file_path = os.path.join(input_folder, file)
        output_file_path = os.path.join(output_folder, os.path.splitext(file)[0] + ".png")
        output_file_path = output_file_path.replace("  ", " ")  # Remove double spaces if any

        # opening message file and extracting the image
        try:
            msg = extract_msg.Message(input_file_path)
            images = msg.attachments

            # saving the extracted image
            if images:
                with open(output_file_path, 'wb') as fp:
                    fp.write(images[0].data)
                print(f"The image has been extracted from {input_file_path} and saved at {output_file_path}.")
            else:
                print(f"No image found in {input_file_path}")
        except Exception as e:
            print(f"Error extracting the image from {input_file_path}: {e}")
```

### cleaning the raw folder, execute after restarting the kernel (process from googledriver blocks deleting files)
